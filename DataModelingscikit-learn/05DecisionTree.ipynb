```python
# Decision Tree
# A decision tree is often a binary tree where nodes 
# represent decisions based on features.
# Nodes are selected based on criteria such as entropy, 
# information gain, or Gini impurity.
# These criteria help determine how effectively a 
# feature splits the data, aiming for the most homogeneous subsets.
```

```python
from sklearn import datasets
from sklearn.tree import DecisionTreeClassifier, DecisionTreeClassifier, plot_tree
import matplotlib.pyplot as plt 

# load dataset
iris = datasets.load_iris()
```

```python
# Initialize and train classifier trees
clf_tree1 = DecisionTreeClassifier()
clf_tree1.fit(iris.data, iris.target)

clf_tree2 = DecisionTreeClassifier(max_depth=8)
clf_tree2.fit(iris.data, iris.target)
```

```python
# helper function to plot a decision tree
def plot_decision_tree(model, features, classes = None):
    plt.figure(figsize=(20,10))
    plot_tree(model,
              filled=True,
              feature_names=features,
              class_names=classes,
              rounded=True,
              proportion=False,
              precision=2)
    plt.show()


plot_decision_tree(clf_tree1, iris.feature_names, iris.target_names)
```

```python
plot_decision_tree(clf_tree2, iris.feature_names, iris.target_names)
```
