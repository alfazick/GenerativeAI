```python
# Grouping Data
# 1. Grouping by column
# 2. Multiple Columns

import pandas as pd

filename = "imaginary_baseball_data.csv"
```

```python
df = pd.read_csv(filename)
print(df)
```

```python
groups = df.groupby('yearID')
for name, group in groups:
    print("Year: {}".format(name))
    print("{}\n".format(group))
```

```python
print('{}\n'.format(groups.get_group(2016)))
```

```python
print('{}\n'.format(groups.sum()))
```

```python
# filtering 
no2015 = groups.filter(lambda x: x.name > 2015) # name here is a name of the group
print(no2015)
```

```python
# 2. Multiple Columns

groups = df.groupby(['yearID','teamID'])
```

```python
for name,group in groups:
    print("Year,Team: {}".format(name))
    print("{}\n".format(group))

```
