```python
# Plotting

# 1. Basics
# 2. Other plots
# 3. Multiple features

import pandas as pd 
import matplotlib.pyplot as plt 
import numpy as np
```

```python
# Not real Data

dates = pd.date_range(start="2023-01-01",periods=10)
# Simultae stock data

data = {
    "Date": np.tile(dates,2), # Repeat dates for each tikcer
    "Ticker":["AAPL"]*10 + ["MSFT"]*10,
    "Open":np.random.uniform(150,200,20),
    "High":np.random.uniform(200,250,20),
    "Low":np.random.uniform(100,150,20),
    "Close":np.random.uniform(150,200,20),
    "Volume":np.random.randint(1e6, 5e6, 20),
}

stocks_df = pd.DataFrame(data)
print(stocks_df)
```

```python

# Assuming stocks_df is your DataFrame and is already defined
msft = stocks_df[stocks_df["Ticker"] == "MSFT"]

# Plotting
msft.plot(kind="line", x="Date", y="High", color="blue")

# Saving the plot before calling plt.show()
plt.savefig("msft_high_prices.png")

# Displaying the plot
plt.show()
```

```python
# Plotting
msft.plot(kind="line", x="Date", y="High", color="blue")

plt.title("Microsoft")
plt.xlabel("January Date")
plt.ylabel("Closing price")
plt.show()
```

```python
# 2. Other plots, "bar", "box"

# Plotting
msft.plot(kind="bar",x= "Date", y="Low")
plt.ylabel("Lowest price")
plt.show()
```

```python
msft.plot(kind="box",y="Open")
plt.show()
```

```python
# Multiple features
# draw multiple features on the same plot


appl = stocks_df[stocks_df["Ticker"] == "AAPL"]

print(msft)
print(appl)
```

```python
ax = plt.gca() # get current axis

msft.plot(kind = "line", x = "Date", y = "Volume", color = "red",ax=ax)

appl.plot(kind='line',x='Date',y='Volume',color = 'blue',ax=ax)

plt.show()
```
