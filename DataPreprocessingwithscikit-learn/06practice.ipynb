```python
# iris data set
# The Iris dataset is ideal for PCA 
# since it contains four features 
# (sepal length, sepal width, petal length, petal width)
# and three classes of iris species
```

```python
import numpy as np
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA 
import matplotlib.pyplot as plt 
```

```python
iris = load_iris()
data = iris.data 
labels = iris.target
label_names = iris.target_names

pca = PCA(n_components=2)
component_data = pca.fit_transform(data)
```

```python
print(labels)
print(label_names)
```

```python
def get_label_info(component_data,labels,class_label,label_names):
    label_name = label_names[class_label]
    label_data = component_data[labels == class_label]
    return (label_name,label_data)

print(get_label_info(component_data,labels,1,label_names))
```

```python
def separate_data(component_data, labels, label_names):
    separated_data = []
    for class_label in range(len(label_names)):
        separated_data.append(get_label_info(component_data,
                                             labels,
                                             class_label,
                                             label_names))
    return separated_data

```

```python
separate_data = separate_data(component_data,labels,label_names)

# plotting the data

for label_name,label_data in separate_data:
    plt.scatter(label_data[:,0],label_data[:,1],label = label_name)

plt.legend()
plt.title('Iris Dataset PCA Plot')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.show()
```
